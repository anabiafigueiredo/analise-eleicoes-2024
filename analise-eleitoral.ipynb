# Análise Exploratória
#####  *Nessa etapa iremos analisar os dados para reposder perguntas relativas as eleições 2024* 

```python
#importando as biibliotecas de visualização dos dados
import seaborn as sns
import matplotlib.pyplot as plt
```

#### Quem é perfil do eleitorado nas eleições municipais em 2024?

```python
df_perfil_eleitorado.columns
```

```python
#A partir desses dados já é possivel identificar 
df_perfil_eleitorado.describe( include = 'all')
```

- Descobrindo a distribuição de gênero do eleitorado

```python
# Calcular a distribuição percentual de gêneros
distribuicao_generos = df_perfil_eleitorado['DS_GENERO'].value_counts(dropna=True)
print(distribuicao_generos)

plt.figure(figsize=(7, 7))  
distribuicao_generos.plot.pie(
    autopct='%1.1f%%',  
    startangle=90,      
    cmap='viridis',     
    labels=distribuicao_generos.index,  
    wedgeprops={'edgecolor': 'black'}   
)


plt.title('Distribuição de Gêneros', loc='left', fontsize=14, pad=20)
plt.axis('equal')
plt.ylabel('')
plt.show()
```

- Visualizando a faixa etária do eleitorado

```python

df_distribuicao_filtrada = df_perfil_eleitorado[df_perfil_eleitorado['DS_FAIXA_ETARIA'] != 'Inválida'] #apagando os dados inválidos
distribuicao_faixa_etaria = df_distribuicao_filtrada['DS_FAIXA_ETARIA'].value_counts().sort_index()
print(distribuicao_faixa_etaria)


plt.figure(figsize=(10, 6))
distribuicao_faixa_etaria.plot(kind='bar', color='purple', edgecolor='black')


plt.xlabel('Faixa Etária')
plt.ylabel('Número de Eleitores')
plt.title('Distribuição de faixa etária do eleitorado')
plt.xticks(rotation=45, ha='right')


plt.tight_layout()
plt.show()
```

- Identificando a raça/cor do eleitorado

```python
df_perfil_eleitorado_filtrado = df_perfil_eleitorado[df_perfil_eleitorado['DS_RACA_COR'] !='NÃO INFORMADO']
distribuicao_raca_cor = df_perfil_eleitorado_filtrado['DS_RACA_COR'].value_counts(normalize=True) * 100
print(distribuicao_raca_cor)

# Criar o gráfico de barras horizontal
plt.figure(figsize=(12, 5))
distribuicao_raca_cor.plot(kind='barh', color='purple', edgecolor='black')

# Adicionar rótulos e título
plt.xlabel('Porcentagem')
plt.ylabel('Raça/Cor')
plt.title('Distribuição de raça/cor do eleitorado (%)')

# Exibir o gráfico
plt.tight_layout()
plt.show()
```

- Quero verificar o grau de escolaridade do eleitorado

```python
distribuicao_grau_escolaridade = df_perfil_eleitorado['DS_GRAU_ESCOLARIDADE'].value_counts(normalize=True) * 100
print(distribuicao_grau_escolaridade)

verde_agua = plt.cm.viridis(0.5)

plt.figure(figsize=(10, 6))
distribuicao_grau_escolaridade.plot(kind='barh', color=verde_agua, edgecolor='black')
plt.xlabel('Porcentagem')
plt.ylabel('Grau de Escolaridade')
plt.title('Distribuição de grau de escolaridade do eleitorado (%)')
plt.tight_layout()
plt.show()
```
